```python
# -*- coding: utf-8 -*-
"""
Created on Sun Oct 24 16:55:59 2021

@author: Ken
"""
import datetime
import numpy as np
from doubleml.datasets import make_plr_CCDDHNR2018
import seaborn as sns
from doubleml import DoubleMLData
from doubleml import DoubleMLPLR
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.base import clone
import joblib
import statsmodels.api as sm



n_rep = 1000

n_obs = 200

n_vars = 150

alpha = 0.5


def d1(x, n_obs, a_0, a_1, s_1):
    return a_0 * x[:, 0] + a_1 * np.divide(np.exp(x[:, 2]), 1 + np.exp(x[:, 2])) \
    + s_1 * np.random.standard_normal(size=[n_obs, ])
    
    
def d2(x, n_obs, a_0, a_1, s_1):
    return a_0 * x[:, 0] + a_1 * np.exp(x[:, 2]) \
    + s_1 * np.random.standard_normal(size=[n_obs, ])
    
def d3(x, n_obs, a_0, a_1, s_1):
    return a_0 * x[:, 0] + a_1 * x[:, 2]**2 \
    + s_1 * np.random.standard_normal(size=[n_obs, ])
    
def d4(x, n_obs, a_0, a_1, s_1):
    return a_0 * x[:, 0] + a_1 * x[:, 2]**3 \
    + s_1 * np.random.standard_normal(size=[n_obs, ])
    
def d5(x, n_obs, a_0, a_1, s_1):
    return a_0 * x[:, 0] + a_1 * x[:, 2] \
    + s_1 * np.random.standard_normal(size=[n_obs, ])
    
def y1(alpha, x, n_obs, b_0, b_1, s_2, d):
    return alpha * d + b_0 * np.divide(np.exp(x[:, 0]), 1 + np.exp(x[:, 0])) \
    + b_1 * x[:, 2] + s_2 * np.random.standard_normal(size=[n_obs, ])

def y2(alpha, x, n_obs, b_0, b_1, s_2, d):
    return alpha * d + b_0 * np.exp(x[:, 0]) \
    + b_1 * x[:, 2] + s_2 * np.random.standard_normal(size=[n_obs, ])

def y3(alpha, x, n_obs, b_0, b_1, s_2, d):
    return alpha * d + b_0 * x[:, 0]**2 \
    + b_1 * x[:, 2] + s_2 * np.random.standard_normal(size=[n_obs, ])

def y4(alpha, x, n_obs, b_0, b_1, s_2, d):
    return alpha * d + b_0 * x[:, 0]**3 \
    + b_1 * x[:, 2] + s_2 * np.random.standard_normal(size=[n_obs, ])
    
def y5(alpha, x, n_obs, b_0, b_1, s_2, d):
    return alpha * d + b_0 * x[:, 0] \
    + b_1 * x[:, 2] + s_2 * np.random.standard_normal(size=[n_obs, ])



#-------------------------------------------------------------------------------
# Data Generation
#-------------------------------------------------------------------------------

def data_gen(fun1, fun2):
    
    data = list()
    
    for i_rep in range(n_rep):
        (x, y, d) = make_plr_CCDDHNR2018(fun1, fun2, alpha=alpha, n_obs=n_obs, dim_x=n_vars, return_type='array')
        data.append((x, y, d))
 
    return data

np.random.seed(1234)

sigmoid_data = data_gen(d1,y1)
exp_data = data_gen(d2,y2)
square_data = data_gen(d3,y3) 
cubic_data = data_gen(d4,y4) 
linear_data = data_gen(d5,y5)
```

```python
cubic_data = list()
for i_rep in range(n_rep):
    (x, y, d) = make_plr_CCDDHNR2018(alpha=alpha, n_obs=n_obs, dim_x=n_vars, return_type='array')
    cubic_data.append((x, y, d))
```

```python

#-------------------------------------------------------------------------------
# OLS Estimation
#-------------------------------------------------------------------------------

def sm_ols(y, X):
    mod = sm.OLS(y, X)
    res = mod.fit()
    res.params
    return res.params, res.conf_int(0.05)

def ols_estimation(data):
    
    theta_ols = np.zeros(n_rep)
    count = []
    
    for i_rep in range(n_rep):
        (x, y, d) = data[i_rep]
        this_theta, conf_i = sm_ols(y, np.column_stack((d, x)))
        theta_ols[i_rep] = this_theta[0]
        count.append(1) if 0.5 >= conf_i[0][0] and 0.5 <= conf_i[0][1] else count.append(0) 
        
    return theta_ols, count


# sigmoid_theta, sigmoid_conf = ols_estimation(sigmoid_data)
# exp_theta, exp_conf = ols_estimation(exp_data)
# square_theta, square_conf = ols_estimation(square_data)
cubic_theta, cubic_conf = ols_estimation(cubic_data)
```

```python

#-------------------------------------------------------------------------------
# Regularized Estimation
#-------------------------------------------------------------------------------

def non_orth_score(y, d, g_hat, m_hat, smpls):
    u_hat = y - g_hat
    psi_a = -np.multiply(d, d)
    psi_b = np.multiply(d, u_hat)
    return psi_a, psi_b




learner = RandomForestRegressor(n_estimators=100, 
                                max_depth=5, 
                                min_samples_leaf=2,
                                n_jobs = -1)

learner2 = XGBRegressor(reg_alpha = 1, reg_lambda = 0)

ml_m = clone(learner2)

ml_g = clone(learner2)

def non_orth_estimation(data):
    
    theta_nonorth = np.zeros(n_rep)
    count = []
    
    for i_rep in range(n_rep):
        (x, y, d) = data[i_rep]
        obj_dml_data = DoubleMLData.from_arrays(x, y, d)
        obj_dml_plr_nonorth = DoubleMLPLR(obj_dml_data,
                                          ml_m, ml_g,
                                          n_folds=2,
                                          draw_sample_splitting=False,
                                          score=non_orth_score)
        obj_dml_plr_nonorth.fit(n_jobs_cv = -1, store_predictions=True)
        this_theta = obj_dml_plr_nonorth.coef[0]
        confint = obj_dml_plr_nonorth.confint()
        count.append(1) if 0.5 <= confint.iloc[0,1] and 0.5 >= confint.iloc[0,0] else 0 
        theta_nonorth[i_rep] = this_theta
        
    return theta_nonorth, count


np.random.seed(1111)

# sigmoid_theta_no, sigmoid_conf_no = non_orth_estimation(sigmoid_data)
# print('first')
# print(datetime.datetime.now())
# exp_theta_no, exp_conf_no     = non_orth_estimation(exp_data)
# print('second')
# print(datetime.datetime.now())
# square_theta_no, square_conf_no  = non_orth_estimation(square_data)
# print('third')
# print(datetime.datetime.now())
cubic_theta_no, cubic_conf_no   = non_orth_estimation(cubic_data)
print('fourth')
```
